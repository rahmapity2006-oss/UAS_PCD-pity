```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import cv2
from PIL import Image, ImageTk
import numpy as np

class AppPCD:
    def __init__(self, root):
        self.root = root
        self.root.title("Fruit Vision Dashboard - Mohammad Sigit")
        self.root.geometry("1200x700")
        
        # --- SIDEBAR (Kiri) ---
        sidebar = tk.Frame(root, bg="#2c3e50", width=250)
        sidebar.pack(side=tk.LEFT, fill=tk.Y)

        tk.Label(sidebar, text="FRUIT ANALYZER", font=('Segoe UI', 16, 'bold'), 
                 bg="#2c3e50", fg="#ecf0f1", pady=20).pack()

        tk.Button(sidebar, text="Load Citra", command=self.load_image, 
                  bg="#27ae60", fg="white", bd=0, height=2).pack(fill=tk.X, padx=20, pady=10)

        ttk.Separator(sidebar, orient='horizontal').pack(fill='x', padx=20, pady=10)

        btn_config = [
            ("Blur (Gaussian)", self.p1_blur), ("Sharpening", self.p2_sharpen),
            ("Edge Detection", self.p3_edge), ("Denoising", self.p4_denoise),
            ("Morfologi (Opening)", self.p5_morfologi), ("Adaptive Thresh", self.p6_adaptive),
            ("Hist. Equalization", self.p7_hist)
        ]
        
        for text, cmd in btn_config:
            tk.Button(sidebar, text=text, command=cmd, bg="#34495e", fg="#ecf0f1", 
                      bd=0, anchor="w", padx=20).pack(fill=tk.X, pady=1)

        ttk.Separator(sidebar, orient='horizontal').pack(fill='x', padx=20, pady=10)
        
        tk.Button(sidebar, text=" SEGMENTASI", command=self.apply_segmentation, bg="#f39c12", fg="white", bd=0, height=2).pack(fill=tk.X, padx=20, pady=5)
        tk.Button(sidebar, text=" EKSTRAKSI FITUR", command=self.extract_features, bg="#e67e22", fg="white", bd=0, height=2).pack(fill=tk.X, padx=20, pady=5)
        tk.Button(sidebar, text=" KLASIFIKASI", command=self.classify_fruit, bg="#c0392b", fg="white", bd=0, height=2).pack(fill=tk.X, padx=20, pady=5)

        # --- MAIN DISPLAY (Kanan) ---
        main_area = tk.Frame(root, bg="#ecf0f1")
        main_area.pack(side=tk.RIGHT, expand=True, fill=tk.BOTH)

        self.panel_asli = tk.Label(main_area, text="Citra Asli", bg="#bdc3c7")
        self.panel_asli.place(relx=0.25, rely=0.4, anchor="center", width=400, height=350)
        
        self.panel_hasil = tk.Label(main_area, text="Citra Hasil", bg="#bdc3c7")
        self.panel_hasil.place(relx=0.75, rely=0.4, anchor="center", width=400, height=350)

        self.img = None
        self.processed_img = None
        self.binary = None

    # --- FUNGSI LOGIKA ---
    def display(self, img, is_original=False):
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) if len(img.shape) == 3 else cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
        img_pil = Image.fromarray(img_rgb).resize((400, 350))
        img_tk = ImageTk.PhotoImage(img_pil)
        if is_original:
            self.panel_asli.config(image=img_tk, text="")
            self.panel_asli.image = img_tk
        else:
            self.panel_hasil.config(image=img_tk, text="")
            self.panel_hasil.image = img_tk

    def load_image(self):
        path = filedialog.askopenfilename()
        if path:
            self.img = cv2.imread(path)
            self.processed_img = self.img.copy()
            self.display(self.img, True)
            self.display(self.img, False)

    def p1_blur(self): self.processed_img = cv2.GaussianBlur(self.img, (5,5), 0); self.display(self.processed_img)
    def p2_sharpen(self): kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]]); self.processed_img = cv2.filter2D(self.img, -1, kernel); self.display(self.processed_img)
    def p3_edge(self): gray = cv2.cvtColor(self.img, cv2.COLOR_BGR2GRAY); self.processed_img = cv2.Canny(gray, 100, 200); self.display(cv2.cvtColor(self.processed_img, cv2.COLOR_GRAY2BGR))
    def p4_denoise(self): self.processed_img = cv2.medianBlur(self.img, 5); self.display(self.processed_img)
    def p5_morfologi(self): kernel = np.ones((5,5), np.uint8); self.processed_img = cv2.morphologyEx(self.img, cv2.MORPH_OPEN, kernel); self.display(self.processed_img)
    def p6_adaptive(self): gray = cv2.cvtColor(self.img, cv2.COLOR_BGR2GRAY); self.processed_img = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2); self.display(cv2.cvtColor(self.processed_img, cv2.COLOR_GRAY2BGR))
    def p7_hist(self): img_yuv = cv2.cvtColor(self.img, cv2.COLOR_BGR2YUV); img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0]); self.processed_img = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR); self.display(self.processed_img)

    def apply_segmentation(self):
        gray = cv2.cvtColor(self.processed_img, cv2.COLOR_BGR2GRAY) if len(self.processed_img.shape) == 3 else self.processed_img
        _, self.binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
        self.display(cv2.cvtColor(self.binary, cv2.COLOR_GRAY2BGR))

    def extract_features(self):
        if self.binary is not None:
            contours, _ = cv2.findContours(self.binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
            if contours:
                c = max(contours, key=cv2.contourArea)
                self.luas = cv2.contourArea(c)
                perimeter = cv2.arcLength(c, True)
                self.circularity = (4 * np.pi * self.luas) / (perimeter ** 2) if perimeter > 0 else 0
                messagebox.showinfo("Fitur", f"Circularity: {self.circularity:.4f}")
        else: messagebox.showwarning("Error", "Lakukan segmentasi!")

    def classify_fruit(self):
        if hasattr(self, 'circularity'):
            if self.circularity > 0.70: hasil = "Jeruk (Bulat)"
            elif 0.35 < self.circularity <= 0.70: hasil = "Apel (Oval)"
            elif 0.05 < self.circularity <= 0.35: hasil = "Pisang (Lonjong)"
            else: hasil = "Tidak Terdeteksi"
            messagebox.showinfo("Hasil", f"Terdeteksi sebagai: {hasil}")
        else: messagebox.showwarning("Error", "Ekstraksi fitur belum dilakukan!")

root = tk.Tk()
app = AppPCD(root)
root.mainloop()
```
